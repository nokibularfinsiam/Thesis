```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, Model, optimizers, callbacks
from tensorflow.keras.applications import DenseNet169, InceptionV3
from tensorflow.keras.applications.densenet import preprocess_input as preprocess_densenet
from tensorflow.keras.applications.inception_v3 import preprocess_input as preprocess_inception
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image

# 1. PARAMETERS
BASE_DIR   = '/content/drive/MyDrive/wrist_xray'
MODELS_DIR = '/content/drive/MyDrive'
TRAIN_DIR  = os.path.join(BASE_DIR, 'train')
VAL_DIR    = os.path.join(BASE_DIR, 'val')
IMG_SIZE   = (299, 299)
BATCH_SIZE = 16
SEED       = 42
W_DENSE    = os.path.join(MODELS_DIR, 'densenet169_wrist_model_safe.h5')
W_INCEPT   = os.path.join(MODELS_DIR, 'inceptionv3_wrist_model_safe.h5')

# 2. BUILD & LOAD FEATURE EXTRACTORS

def build_feature_model(base_cls, gap_name):
    inp = layers.Input((*IMG_SIZE, 3))
    base = base_cls(weights='imagenet', include_top=False, input_tensor=inp)
    base.trainable = False
    x = layers.GlobalAveragePooling2D(name=gap_name)(base.output)
    return Model(inp, x)

# DenseNet169 and InceptionV3 feature extractors
densenet_feat = build_feature_model(DenseNet169, 'dn_gap')
densenet_feat.load_weights(W_DENSE, by_name=True, skip_mismatch=True)

inception_feat = build_feature_model(InceptionV3, 'in_gap')
inception_feat.load_weights(W_INCEPT, by_name=True, skip_mismatch=True)

# 3. TF.DATA PIPELINE

def make_dataset(directory, training=True):
    ds = tf.keras.preprocessing.image_dataset_from_directory(
        directory,
        labels='inferred', label_mode='binary',
        image_size=IMG_SIZE, batch_size=BATCH_SIZE,
        shuffle=training, seed=SEED
    )
    # Normalize and augment for training
    if training:
        ds = ds.map(lambda x, y: (tf.image.random_flip_left_right(x), y), num_parallel_calls=tf.data.AUTOTUNE)
        ds = ds.map(lambda x, y: (tf.image.random_brightness(x, 0.2), y), num_parallel_calls=tf.data.AUTOTUNE)
    # Dual preprocess for both branches
    ds = ds.map(
        lambda x, y: ((preprocess_inception(x), preprocess_densenet(x)), y),
        num_parallel_calls=tf.data.AUTOTUNE
    )
    return ds.prefetch(tf.data.AUTOTUNE)

train_ds = make_dataset(TRAIN_DIR, training=True)
val_ds   = make_dataset(VAL_DIR, training=False)

# 4. BUILD ENSEMBLE HEAD

inp_in = layers.Input((*IMG_SIZE,3), name='input_in')
inp_dn = layers.Input((*IMG_SIZE,3), name='input_dn')
feat_i = inception_feat(inp_in)
feat_d = densenet_feat(inp_dn)
merged = layers.Concatenate()([feat_i, feat_d])

x = merged
for units, drop_rate in [(512, .5), (256, .4), (128, .3)]:
    x = layers.BatchNormalization()(x)
    x = layers.Dense(units, activation='relu')(x)
    x = layers.Dropout(drop_rate)(x)
out = layers.Dense(1, activation='sigmoid')(x)

wristnet = Model([inp_in, inp_dn], out)

# 5. COMPILE
wristnet.compile(
    optimizer=optimizers.Adam(1e-4),
    loss='binary_crossentropy',
    metrics=['accuracy',
             tf.keras.metrics.Precision(name='precision'),
             tf.keras.metrics.Recall(name='recall'),
             tf.keras.metrics.AUC(name='auc')]
)

# 6. CALLBACKS
cbs = [
    callbacks.EarlyStopping('val_accuracy', patience=8, restore_best_weights=True),
    callbacks.ModelCheckpoint('wristnet_best.h5', save_best_only=True, monitor='val_accuracy'),
    callbacks.ReduceLROnPlateau('val_loss', factor=0.2, patience=4)
]

# 7. TRAIN
history = wristnet.fit(
    train_ds,
    validation_data=val_ds,
    epochs=30,
    callbacks=cbs
)

# 8. EVALUATE & SAVE
wristnet.load_weights('wristnet_best.h5')
evals = wristnet.evaluate(val_ds)
print(f"Accuracy: {evals[1]*100:.2f}% | Precision: {evals[2]*100:.2f}% | Recall: {evals[3]*100:.2f}% | AUC: {evals[4]*100:.2f}%")
wristnet.save('wristnet_final.h5')

# 9. PLOT HISTORY
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='train acc')
plt.plot(history.history['val_accuracy'], label='val acc')
plt.title('Accuracy'); plt.legend(); plt.ylim(0,1)
plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='train loss')
plt.plot(history.history['val_loss'], label='val loss')
plt.title('Loss'); plt.legend()
plt.tight_layout(); plt.show()

# 10. PREDICTION UTILITY
def predict_wristnet(img_path, model):
    img = image.load_img(img_path, target_size=IMG_SIZE)
    arr = np.array(img)
    arr_i = preprocess_inception(arr)
    arr_d = preprocess_densenet(arr)
    pred = model.predict([[np.expand_dims(arr_i,0)], [np.expand_dims(arr_d,0)]])[0][0]
    label = 'Fracture' if pred < 0.5 else 'Normal'
    color = 'red' if label=='Fracture' else 'green'
    plt.imshow(img)
    plt.title(label, color=color)
    plt.axis('off')
    plt.show()
```
